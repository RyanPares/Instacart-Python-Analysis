# 4. Create a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Aggregating Data with transform() Function
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking the Output
df.head(15)
```

```python
# Creating a 'Loyalty' Label Flag using loc Function
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Examine Value Counts on 'loyalty_flag'
df['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking Updated Output
df.head(15)
```

# 5. Check if there’s a difference between the spending habits of the 3 types of customers (Loyal Customer, Regular Customer, and New Customer)

```python
# Determining Basic Statistics of Spending Habits by Loyalty Category
df.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

#### Loyal customers generally opt for more affordable purchases, averaging 10.39 per item. In contrast, new customers appear to spend more freely, with their average expenditure per item being 13.29. Regular customers' spending habits are intermediate, with an average spend of 12.50 per item. While the minimum spend across all customer types is 1.0, the presence of a maximum spend of 99,999 could indicate the presence of outliers or inaccuracies within the data.

# 6. Create a spending flag for each user based on the average price across all their orders

```python
# Aggregating Data with transform() Function
df['avg_product_price'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking Relevant Columns 
df[['user_id', 'avg_product_price']].head(15)
```

```python
# Creating a 'Spending' Label Flag using loc Function
df.loc[df['avg_product_price'] >= 10, 'spending_flag'] = 'High spender'
df.loc[df['avg_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Examine Value Count on 'spending_flag' Column
df['spending_flag'].value_counts(dropna = False)
```

```python
# Checking Releavent Columns
df[['user_id', 'avg_product_price', 'spending_flag']].head(15)
```

# 7. Determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order”

```python
# Aggregating Data with transform() Function
df['median_days_between_orders'] = df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking Relevant Columns 
df[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(15)
```

```python
# Creating a 'Order Frequency' Label Flag using loc Function
df.loc[df['median_days_between_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df.loc[(df['median_days_between_orders'] <= 20) & (df['median_days_between_orders'] > 10), 'order_frequency_flag'] = 'Regular customer'
df.loc[df['median_days_between_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Examine Value Counts on 'order_frequency_flag'
df['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Checking Relevant Columns of Updated Dataframe
df[['user_id', 'days_since_prior_order', 'median_days_between_orders', 'order_frequency_flag']].head(15)
```
